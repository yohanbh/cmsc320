Due to the nature of this assignment, we will be working with relatively recent eruptions, although some are still older than anyone currently alive on Earth. Each eruption in the database is classified as significant if it meets one or more criteria, such as causing fatalities, causing **property damage** (**more than $1 million**), reaching a **Volcanic Explosivity Index (VEI)** of **6 or higher**, generating a tsunami, or being linked to a significant earthquake. The VEI is a scale that measures the explosiveness of volcanic eruptions, providing insight into their magnitude and potential consequences. The database includes detailed information on the location, volcano type, last known eruption, VEI, casualties, property damage, and much more.
